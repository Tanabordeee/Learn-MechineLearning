```python
import pandas as pd
df = pd.read_excel("Stock.xlsx", index_col="Name")
df
```

```python
# สั่งให้เรียง index โดย default เรียงจากน้อยไปมาก

result = df.sort_index()
result
```

```python
# ถ้าเรียงจากมากไปน้อย df.sort_index(ascending=False)
# inplace = True ก็จะทำให้มัน save ทับ dataFrame
result2 = df.sort_index(ascending=False , inplace=True)
result2
df
```

```python
# เรียงตัว values
# โดย defaults จะเรียงจากน้อยไปมาก
result_sort = df.sort_values("Price")
result_sort
```

```python
# เรืยง values จากมากไปน้อย 
# แล้วถ้าอยากให้มัน save ทับลงไปใน dataFrame เลยก็ inplace = True
df.sort_values("Price" , ascending=False , inplace=True)
df
```

```python
# เพิ่ม columns
df["delivery"] = 100
df
```

```python
#  เพิ่ม columns โดย ใส่ values เป็นค่าของ 2 columns บวกกัน
df["total"] = df["Price"] + df["delivery"]
df
```

```python
# การเปลี่ยน ชื่อ column แค่ column เดียว
# ให้สร้าง dict โดย {"ชื่อเดิม" : "ชื่อใหม่"}
cols = {"Amount":"ffff"}
# set ให้ inplace = True เพื่อเขียนทับลง dataFrame ไปเลย
df.rename(columns=cols , inplace=True)
df
```

```python
# ใช้ในการเปลี่ยนชื่อ columns หลายๆอันพร้อมกัน
cols = {"Category" : "Group" , "ffff":"Price" , "total":"summation"}
df.rename(columns=cols , inplace=True)
df
```

```python
# การลบ columns โดยมันจะลบแต่ไม่ save ทับลง dataFrame ไม่ได้ลบ ถาวร
# axis = 1 => columns
# axis = 0 => row ตาม index ที่ระบุ
df.drop("delivery" , axis=1)
```

```python
# ลบ columns แบบ ถาวร save ทับลงไปใน dataFrame
df.drop("delivery" , axis=1 , inplace=True)
df
```
