```python
import pandas as pd
df = pd.read_csv("pima_diab.csv")
df.head(20)
```

```python
# แสดงจำนวนจากการนับจำนวนแถว
df.groupby("class").size()
```

```python
df.drop("Unnamed: 0", axis=1 , inplace=True)
df
```

```python
df.fillna(df.mean(numeric_only=True) , inplace=True)
df
```

```python
df["class"].fillna(df["class"].mode()[0] , inplace=True)
df
```

```python
df["class"] = df["class"].replace({
    "tested_negative" : 0 ,
    "tested_positive": 1
})
df
```

```python
new_df = df[df.corr()["class"].drop("class").sort_values(ascending=False).head(4).keys()]
new_df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(new_df , df["class"] , test_size=0.3 , random_state=0)
print(x_test.shape)
print(x_train.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train , y_train)
model.score(x_test , y_test)

```

```python
lables = model.predict(new_df.loc[[2]])
print(lables)
```

```python
# หา test แถวที่ 3
test_sample = x_test.iloc[[3]]

# หา 5 neighbors ที่ใกล้ที่สุด
distance , indices = model.kneighbors(test_sample, n_neighbors=5)
print(distance)

# ตำแหน่งของ neighbors ใน training set
print(indices)

# indices[0] คือ list ของตำแหน่ง (index) training samples 5 ตัวที่ใกล้ที่สุด
nest_sample = x_train.iloc[indices[0]]
nest_sample 
```

```python
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
parameter = {"n_neighbors":list(range(1,21))}
clf = GridSearchCV(knn ,parameter)
clf.fit(x_train , y_train)
clf.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(x_train , y_train)
knn.score(x_test , y_test)
```

```python
import pickle 
pickle.dump(knn , open("knn.pkl" , "wb"))
```
